**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Create a function that calculates the ratio of 2018 average price to 1998 average price for inputted df: 
def create_price_ratio(d): 
    # Extract average prices for 1998 and 2018
    year1998 = float(d.loc[d['Year']==1998, 'Average_price'].iloc[0])
    year2018 = float(d.loc[d['Year']==2018, 'Average_price'].iloc[0])
    
    # Calculate the ratio
    ratio = year2018 / year1998
    return ratio
```

```python
# Test out the function: 
create_price_ratio(dfg[dfg['London_Borough']=='Camden'])
```

```python
# Calculate for all Boroughs, so first make an empty dictionary to store ratios for each unique Borough
final = {}
```

```python
# For loop to iterate through each unique borough
for b in dfg['London_Borough'].unique():
    # Make a parameter for create_price_ratio function: subsetting dfg on 'London_Borough' == b. 
    borough = dfg[dfg['London_Borough'] == b]
    # New entry in the final dictionary w/ result value from calling function w/ argument: borough
    final[b] = create_price_ratio(borough)
# Print
print(final) 
```

```python
# Make a variable and assign it the result of calling the DataFrame method on the dictionary final
# Create a Series from the dictionary
series_data = pd.Series(final, name='Price_Ratio')

# Create a DataFrame by concatenating along columns
df_ratios = pd.DataFrame(series_data).transpose()

# Print the DataFrame
print(df_ratios)
```

```python
# Call head() method to check it out
df_ratios.head()
```

```python
# Transpose df and reset index 
df_ratios_T = df_ratios.transpose()
df_ratios = df_ratios_T.reset_index()
df_ratios.head()
```

```python
# Rename the 'index' column as 'London_Borough'
df_ratios.rename(columns={'index':'Borough'}, inplace=True)
df_ratios.head()
```

```python
# Sort in descending order and select the top 15 boroughs.
top15 = df_ratios.sort_values(by='Price_Ratio',ascending=False).head(15)
print(top15)
```

```python
# Plot the boroughs w/ greatest price change
ax = top15[['Borough','Price_Ratio']].plot(kind='bar')

ax.set_xticklabels(top15.Borough)
```
